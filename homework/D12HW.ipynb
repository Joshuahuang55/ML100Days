```python
import pandas as pd
import numpy as np
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
boston = pd.read_csv("boston_D12.csv")
boston.boxplot()
print("TAX_median: ",boston.TAX.median())
print("B_median: ",boston.B.median())
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
data = {"NOX": boston.NOX , "DIS" : boston.DIS}
df = pd.DataFrame(data)
df.plot.scatter(x = "NOX" , y = "DIS", title = "Negative correlation")
```
